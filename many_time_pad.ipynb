## XOR phase

The assignment asks us to XOR the cyphers together.
From my understanding, if we have plaintext m with key k chosen from keyset K, we have the following:
>> m XOR k

with this, if we XOR the message with k again, we simply get m.
However; if there are multiple messages with the same key used, if we XOR each one of them, we remove k and instead we are left
with the XOR of message m1, m2 for any two messages.
Then, we can deduce the contents of the message with frequency analysis.

## HINT

if a space is XORed with a character:  
`Space in ascii corresponds to hex value '20', or binary '00100000'.`  
`Capital A in ascii is:                  '41', binary:   '01010000'.`  
Let's do a loop and see what the actual values will be.  


```python
space = int('20', 16)
print("spa" + format(space, '< 16b'))

# list comprehension is unnecesary since these values can be represented as ints already.
# however, it's helpful for exploration.
uppercase_letters = range(65, 90)
lowercase_letters = range(97, 122)

# show the original value, and then the value after it's been xor'ed with space.
print("uppercase letters")
for letter in uppercase_letters:
    print(format(letter) + format(letter, '< 16b') + format(letter ^ space, '< 16b') + format(letter ^ space))

print("spa" + format(space, '< 16b'))
print("lowercase letters")
for letter in lowercase_letters:
    print(format(letter) + format(letter, '< 16b') + format(letter ^ space, '< 16b') + format(letter ^ space))
```

This is important, because space in ints is 32 - the difference between the lower case and upper case in ascii is  
also 32. What this means is this shows the modulo 2 sum capability of XOR bitwise.
